**Introduction/Business Problem.**

The pandemic due to COVID-19 is ending in my country, Uruguay. After a very effective vaccination plan, the daily cases of contagion are less and less.
 
This means that businesses are reopening to the public who, after several months, want to return to those meetings with their friends, and the social life that characterizes us Uruguayans so much. Considering all this, won't this be the best time to open a new place? People talk a lot about the “new normal”, this will also mean new restaurants, new bars, new shopping centers, where people will go to have a good time. The question that arises is, what kind of place should I open?

This work will answer that question from data.

For this, the use of Foursquare will be essential. Foursquare is a web location-based service applied to social networks. Through this application, we can find all the businesses open anywhere in the world, and not only that, but also what type of place they are, reviews, and even how often they are visited.
Using the API of this application, we will analyze each neighborhood of Montevideo, Uruguay, finding the most popular types of business in each of the localities, and thus, try to answer the question that this Data Science study tries to answer.


**Data Section**

The data used in this project can be divided into two main parts.
- Database on Montevideo:
The first job to do was to get a database with all the neighborhoods of the capital of Uruguay, and their geolocation. This work was not easy, since the information was not in a single web page. Finally, after carrying out the research work, the database was completed, with all the postal codes, neighborhoods, latitudes and longitudes.
 
- Foursquare:
The application database will be the main part of our work. Thanks to the API, we will be able to access all the information that you mention in the Introduction.
Going more specifically, we are going to use:

    1) Location of places registered in the application.
    2) Type of business.
    3) Number of these businesses in each of the neighborhoods.
    4) Frequency with which they are visited.
